```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_brandenburg = pd.read_csv('../Data/brandenburg_cleaned.csv')
df_brandenburg.head()
```

```python
df_brandenburg.columns
```

```python
# Create a loop to generate main factors based on cities in Brandenburg
vars_of_interest = ['city', 'balcony','lift', 'pets_allowed', 'floor',
       'no_rooms', 'garden', 'cellar', 'noParkSpaces']

for var in vars_of_interest:
    plt.figure(figsize=[15,5])
    sns.countplot(data = df_brandenburg, x=var)
    plt.xticks(rotation=30)
```

```python
# get a view of the total rent prices pro city in Brandenburg
# Total Rent price distribution in Land Brandenburg (Germany)
plt.figure(figsize=[20,6])
city_order = list(df_brandenburg.groupby(['city']).total_rent.mean().sort_values().index)
sns.barplot(data=df_brandenburg, x='city', y='total_rent', order=city_order)
plt.xticks(rotation=90);
```

```python
# get a view of the base rent prices pro city in Brandenburg
# Base Rent price distribution in Land Brandenburg (Germany)
plt.figure(figsize=[20,6])
city_order = list(df_brandenburg.groupby(['city']).total_rent.mean().sort_values().index)
sns.barplot(data=df_brandenburg, x='city', y='base_rent', order=city_order)
plt.xticks(rotation=90);
```

### 2.Which city has the most expensive and cheapest rent in Land Brandenburg (Germany)? 
#### Potsdam(the most expensive) and Prignitz_Kreis(the cheapest)

```python
df_brandenburg.groupby(['city']).total_rent.mean().sort_values(ascending=False)
```

## Correlations

```python
plt.figure(figsize=(10, 10))

numData = df_brandenburg._get_numeric_data()
var_num_corr = numData.corr()

sns.heatmap(var_num_corr, vmin=-1, vmax=1, annot=True, linewidth=0.01, linecolor='black', cmap='RdBu_r')

plt.show()
```

```python
var_num_corr['total_rent'].round(3)
```

I can accept that the factors that most positively influence the rent amount (correlation> = 0.3) are: living space, no_rooms,noParkSpaces are clearly correlated

## 3. Analysing of each important factors

### Main Factors which has an impact on total_rent price in Brandenburg

### living_space

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.regplot(x=df_brandenburg['living_space'], y=df_brandenburg['total_rent'], line_kws={'color': 'r'})
plt.xscale('log')
plt.yscale('log')
plt.savefig("plot.png", dpi=300)
plt.show()
```

he number of rooms varies between 1 and 5, and I noticed that the more rooms, the higher the rent, which is already expected.y

### rooms

```python
plt.figure(figsize=(12, 6))
plt.subplot()
sns.barplot(x=df_brandenburg['no_rooms'], y=df_brandenburg['total_rent'])
plt.savefig("plot.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.regplot(x=df_brandenburg['no_rooms'], y=df_brandenburg['total_rent'], line_kws={'color': 'r'})
plt.xscale('log')
plt.yscale('log')
plt.savefig("plot.png", dpi=300)
plt.show()
```

The number of rooms varies between 1 and 5, and I noticed that the more rooms, the higher the rent, which is already expected.

### ParkSpaces

```python
## impact of number of park spaces in the rent price
plt.figure(figsize=(12, 6))
plt.subplot()
sns.barplot(x=df_brandenburg['noParkSpaces'], y=df_brandenburg['total_rent'])
plt.savefig("plot.png", dpi=300)
plt.show()
```

The number of parking spaces usually varies between 0 and 2, and we noticed that the more parking spaces, the higher the rent, which is already expected.
